```python
def statetax(input, money, quant):
    if "DC" in input.upper():
        money=money*0.053*quant
    elif "MD" in input.upper():
        money=money*0.06*quant
    elif "VA" in input.upper():
        money=money*0.0575*quant
    else:
        money=money*0.05*quant
    return money
```

```python
def evenspacing(var):
    """Make the spacing the same."""
    length=len(var)
    while length<10:
        var=var+" "
        length=length+1
    return var
```

```python
name = input("Enter your name: ")
address = input ("Enter you address: ")
date = input("Enter the date: ")
account = input("Enter you account number: ")

invoice={}

while True:
    item = input("\nEnter item name: ")
    price = input("Price: ")
    price=float(price)
    quantity = input("Quantity: ")
    quantity = int(quantity)
    taxable = input("Taxable (true or false): ")
    if "true" in taxable.lower():
        state = input("Enter your state: ")
        cost = statetax(state, price, quantity)
    else:
        cost = 0
    order = {"Quantity":quantity, "Price":price, "Cost":cost, "Taxable":taxable}
    invoice[item] = order
    cont = input("Add another item (y/n): ")
    if "n" in cont.lower():
        break

print("\n-----------------------------------------------------")       
print("Customer Name: " + name.title())
print("Address: " + address)
print("Date: " + date)
print("Account Number: " + str(account))
print("-----------------------------------------------------\n")
print("Item Name\tQuantity\tPrice\tCost\tTaxable")
print("=========================================================================")
subtotal=0
taxes=0
for food, things in invoice.items():
    qua=things["Quantity"]
    pri=things["Price"]
    costy=things["Cost"]
    taxy=things["Taxable"]
    food=evenspacing(food)
    print(food + "\t" + str(qua) + "\t\t" + str("{:.2f}".format(pri,2)) + "\t" + str("{:.2f}".format(qua*pri,2)) + "\t" + taxy)
    subtotal=subtotal + (qua*pri)
    taxes=taxes + costy
print("\nSubtotal: $" + str("{:.2f}".format(subtotal,2)))
print("Sales Tax: $" + str("{:.2f}".format(taxes,2)))
print("\n-------------------------------------------------------------------------")
print("Total: $" + str("{:.2f}".format(subtotal+taxes,2)))
```
