```python
import random

nosell = "You can't go that way!"

rooms={1:{'name':'foyer','north':2,'east':0,'west':0,'south':0,'object':'dead scorpion'},
            2:{'name':'front room','north':0,'east':4,'west':3,'south':1,'object':'piano'},
            3:{'name':'library','north':5,'east':2,'west':0,'south':0,'object':'spiders'},
            4:{'name':'kitchen','north':7,'east':0,'west':2,'south':0,'object':'bats'},
            5:{'name':'dining room','north':0,'east':0,'west':0,'south':3,'object':'dust and an empty box'},
            6:{'name':'vault','north':0,'east':8,'west':0,'south':0,'object':'3 walking skeletons'},
            7:{'name':'parlor','north':0,'east':0,'west':6,'south':4,'object':'treasure chest'},
            8:{'name':'secret room','north':0,'east':0,'west':6,'south':0,'object':'piles of gold'},
        }
```

```python
def walking(var):
    if var == 1:
        print("You can go north.")
    elif var == 2:
        print("You can go south, west, or east.")
    elif var == 3:
        print("You can go north or east.")
    elif var == 4:
        print("You can go north or west.")
    elif var == 5:
        print("You can go south.")
    elif var == 6:
        print("You can go east.")
    elif var == 7:
        print("You can go south or west.")
    else:
        print("You can go west.")
```

```python
def countrooms(var,places):
    if var in places:
        places.remove(var)
    return places
```

```python
place = 1
x=0
roomnums=[1,2,3,4,5,6,7,8]

while True:
    print("\nYou are in the " + rooms[place]['name'] + ". What's inside: " + rooms[place]['object'])
    walking(place)
    countrooms(place,roomnums)
    direction = input("Type a direction: ")
    if direction.lower() == 'q':
        break
    elif direction.lower()== 'north' or direction.lower()== 'south' or direction.lower()== 'west' or direction.lower()== 'east':
        newplace = rooms[place][direction.lower()]
        if newplace == 0:
            print(nosell)
        elif newplace==8:
            if x==0 and random.randint(1,4) != 1:
                place = 7
            else:
                place = 8
                x=1
        else:
            place = newplace
    else:
        print("I don't understand " + direction)
print("\nYou found " + str(8-len(roomnums)) + " rooms!")
if random.randint(1,4) == 1:
    print("You were followed by a ghost.")
```
