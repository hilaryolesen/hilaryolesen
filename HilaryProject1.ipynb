```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
moviedf=pd.read_csv('boxoffice.csv')
moviedf.head()
```

```python
moviedf.set_index('rank', inplace=True)
moviedf.head()
```

```python
movie=moviedf.dropna()
movie.head()
```

```python
movie=movie.drop_duplicates()
movie.head()
```

```python
years = [2014,2015,2016,2017,2018,2019]
movie.year.isin(years)
movie5=movie[movie.year.isin(years)]
movie5.tail()
```

```python
movie5['studio'].unique()
```

```python
temp=movie5.groupby('studio')['title'].nunique()
temp.sort_values()
```

```python
replacements = {
  'WB (NL)': 'WB',
  'FoxS': 'Fox',
}
movie5['studio'].replace(replacements, inplace=True)
```

```python
movie5 = movie5[movie5.studio != 'Blue Fox']
```

```python
list=["IFC","Magn.","KL","Uni.","WGUSA","WB","Fox","SPC","Strand","Sony"]

mostmovie5=movie5[movie5.studio.str.contains('|'.join(list))]

mostmovie5.head(30)
```

```python
moviemelt=pd.melt(mostmovie5,id_vars=['title','studio','year'],var_name="gross")
moviemelt.head()
```

```python
moviemelt['gross'].replace("lifetime_gross", "movies made from 2014-2018", inplace=True)
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='gross',y='value',data=moviemelt,hue='studio',dodge=True)

plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.title('Amount Grossed Per Movie for the 10 Studios with Highest Rate of Production')
```

```python
newlist=["IFC","Magn.","KL","Uni.","WGUSA","WB","Fox","SPC","Strand","Sony","BV"]

newmostmovie5=movie5[movie5.studio.str.contains('|'.join(newlist))]

newmostmovie5.head()
```

```python
newmoviemelt=pd.melt(newmostmovie5,id_vars=['title','studio','year'],var_name="gross")
newmoviemelt.head()
```

```python
newmoviemelt['gross'].replace("lifetime_gross", "movies made from 2014-2018", inplace=True)
```

```python
plt.figure(figsize=(10,6))

sns.swarmplot(x='gross',y='value',data=newmoviemelt,hue='studio',dodge=True)

plt.legend(bbox_to_anchor=(1,1), loc=2)

plt.title('Amount Grossed Per Movie for the 10 Studios with Highest Rate of Production (+Disney)')
```
